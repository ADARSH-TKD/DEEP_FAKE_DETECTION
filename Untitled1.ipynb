```python
# --- CONFIGURATION ---
DATASET_PATH = r'C:\Users\adarsh\OneDrive\Documents\GitHub\DEEP_FAKE_DETECTION\data_set'  # Update this path
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 20
MODEL_NAME = 'best_simple_cnn_deepfake_model.h5'

# Define the data generator function (from original script, but simplified imports)
def create_data_generators(dataset_path, img_size=IMG_SIZE, batch_size=BATCH_SIZE):
    # Data augmentation for training (rescaling, rotation, shifts, flips, zoom, shear)
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True,
        zoom_range=0.2,
        shear_range=0.2,
        fill_mode='nearest'
    )

    # Only rescaling for validation and test (no augmentation)
    val_test_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(
        os.path.join(dataset_path, 'Train'),
        target_size=img_size,
        batch_size=batch_size,
        class_mode='binary',
        classes=['Real', 'Fake'],
        shuffle=True
    )

    validation_generator = val_test_datagen.flow_from_directory(
        os.path.join(dataset_path, 'Validation'),
        target_size=img_size,
        batch_size=batch_size,
        class_mode='binary',
        classes=['Real', 'Fake'],
        shuffle=False
    )

    test_generator = val_test_datagen.flow_from_directory(
        os.path.join(dataset_path, 'Test'),
        target_size=img_size,
        batch_size=batch_size,
        class_mode='binary',
        classes=['Real', 'Fake'],
        shuffle=False
    )

    return train_generator, validation_generator, test_generator

# Execute generator creation
print("[STEP 1] Creating Data Generators...")
train_gen, val_gen, test_gen = create_data_generators(DATASET_PATH)

print(f"\n✓ Data generators created successfully!")
print(f"  Training samples: {train_gen.samples}")
print(f"  Validation samples: {val_gen.samples}")
print(f"  Test samples: {test_gen.samples}")
print(f"  Classes: {train_gen.class_indices}")
```

```python
## 📊 Visualize Dataset and Pre-processing

# Function to display images and their labels
def visualize_samples(generator, num_samples=5):
    class_labels = list(generator.class_indices.keys())
    batch_images, batch_labels = next(generator)
    
    plt.figure(figsize=(15, 5))
    for i in range(num_samples):
        ax = plt.subplot(1, num_samples, i + 1)
        
        # Display image. Image data is already rescaled [0, 1]
        plt.imshow(batch_images[i])
        
        # Binary label (0 or 1) to string label ('Real' or 'Fake')
        label_index = int(batch_labels[i])
        label = class_labels[label_index]
        
        ax.set_title(f"Label: {label}", fontsize=10)
        ax.axis("off")
    plt.suptitle("Sample Images from Training Set (Pre-processed/Augmented)", fontsize=16, fontweight='bold')
    plt.show()

# Visualize 5 samples from the training generator
print("\n[STEP 3.1] Visualizing 5 Sample Images (with augmentation/rescaling)...")
visualize_samples(train_gen)

# Show what augmentation does to a single image (optional, but helpful)
print("\n[STEP 3.2] Visualizing Augmentation on a Single Image...")

# Temporarily create a generator with flow_from_directory on a single image
temp_datagen = ImageDataGenerator(
    rescale=1./255, rotation_range=40, width_shift_range=0.2,
    height_shift_range=0.2, shear_range=0.2, zoom_range=0.2, horizontal_flip=True
)

# Find a representative image path (e.g., the first 'Real' image)
sample_dir = os.path.join(DATASET_PATH, 'Train', 'Real')
sample_img_path = os.path.join(sample_dir, os.listdir(sample_dir)[0])

img = keras.utils.load_img(sample_img_path, target_size=IMG_SIZE)
x = keras.utils.img_to_array(img)
x = x.reshape((1,) + x.shape) / 255.0 # Reshape and normalize

plt.figure(figsize=(10, 5))
i = 0
for batch in temp_datagen.flow(x, batch_size=1):
    plt.subplot(1, 5, i + 1)
    plt.imshow(keras.utils.img_to_array(batch[0]))
    plt.axis("off")
    i += 1
    if i % 5 == 0:
        break
plt.suptitle("Augmented Versions of a Single Image", fontsize=14, fontweight='bold')
plt.show()
```

```python
## 🧱 Revised Step 4: Build Simpler CNN Model

# --- Configuration from previous steps ---
IMG_SIZE = (224, 224)
MODEL_NAME = 'best_simpler_cnn_deepfake_model_v2.h5' # Changed model name to save new version

def build_simpler_cnn(input_shape=(*IMG_SIZE, 3)):
    """
    Simpler CNN Model with increased regularization (Dropout) 
    to prevent overfitting on the small dataset.
    """
    model = models.Sequential([
        # Block 1
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        
        # Block 2
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.3), 
        
        # Block 3
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.4),
        
        # Flatten and Dense Layers
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.6), # Increased Dropout
        
        # Output Layer
        layers.Dense(1, activation='sigmoid')
    ])
    
    return model

print("\n[STEP 4] Building REVISED Pure CNN Model...")
model = build_simpler_cnn(input_shape=(*IMG_SIZE, 3))
print("✓ Simpler, More Regularized CNN Model created")
    
print(f"\nModel Summary:")
print(f"  Total parameters: {model.count_params():,}")

# Compile model: Using a smaller learning rate to stabilize training
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0001), # Reduced learning rate
    loss='binary_crossentropy',
    metrics=['accuracy', 
             keras.metrics.Precision(name='precision'),
             keras.metrics.Recall(name='recall'),
             keras.metrics.AUC(name='auc')]
)
```

```python
## 🚀 Revised Step 5: Train Model

# --- Configuration from previous steps ---
EPOCHS = 20
MODEL_NAME = 'best_simpler_cnn_deepfake_model_v2.h5' 
# Assuming train_gen and val_gen were created in the original Step 2

def train_model(model, train_gen, val_gen, epochs=EPOCHS, model_name=MODEL_NAME):
    """
    Train the model using generators with revised callbacks.
    """
    # Callbacks
    callbacks = [
        # Save best model
        keras.callbacks.ModelCheckpoint(
            model_name,
            save_best_only=True,
            monitor='val_accuracy',
            mode='max',
            verbose=1
        ),
        # Reduce learning rate when plateauing (patience remains 3)
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-7,
            verbose=1
        ),
        # Early stopping (Patience increased to 10)
        keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10, # Increased patience!
            restore_best_weights=True,
            verbose=1
        ),
    ]
    
    print("\n" + "="*60)
    print("STARTING REVISED TRAINING")
    print("="*60)
    
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=epochs,
        callbacks=callbacks,
        verbose=1
    )
    
    return history

print("\n[STEP 5] Training Revised Model...")
# Ensure you use the generators created in the original Step 2
history = train_model(model, train_gen, val_gen)
```

```python
## 📈 Plot Training History

def plot_training_history(history, save_path='cnn_training_history.png'):
    """Plot and save training metrics"""
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    metrics = ['accuracy', 'loss', 'precision', 'recall']
    titles = ['Model Accuracy', 'Model Loss', 'Model Precision', 'Model Recall']
    
    for i, metric in enumerate(metrics):
        row = i // 2
        col = i % 2
        
        axes[row, col].plot(history.history[metric], label='Train', linewidth=2)
        axes[row, col].plot(history.history[f'val_{metric}'], label='Validation', linewidth=2)
        axes[row, col].set_title(titles[i], fontsize=14, fontweight='bold')
        axes[row, col].set_xlabel('Epoch')
        axes[row, col].set_ylabel(metric.capitalize())
        axes[row, col].legend()
        axes[row, col].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    print(f"\n✓ Training history saved to {save_path}")
    plt.show()

print("\n[STEP 6] Plotting Training Results...")
plot_training_history(history)
```

```python
## 🧪 Final Evaluation & Confusion Matrix

# Load the best model weights for evaluation
try:
    best_model = load_model(MODEL_NAME)
except:
    print(f"Warning: Could not load {MODEL_NAME}. Using the model from the last epoch.")
    best_model = model


print("\n[STEP 7.1] Evaluating Model on Test Set...")
print("="*60)
print("TEST SET RESULTS")
print("="*60)

test_results = best_model.evaluate(test_gen, verbose=1)
metric_names = ['Loss', 'Accuracy', 'Precision', 'Recall', 'AUC']

print("\n" + "="*60)
print("TEST SET SUMMARY")
print("="*60)
for name, result in zip(metric_names, test_results):
    print(f"{name}: {result:.4f} ({result*100:.2f}%)" if name == 'Accuracy' else f"{name}: {result:.4f}")
print("="*60)


# --- Confusion Matrix ---
print("\n[STEP 7.2] Generating Confusion Matrix...")
# Predict probabilities on the test set
test_gen.reset() # Reset generator to ensure predictions match labels
predictions = best_model.predict(test_gen, verbose=1)
y_pred_classes = (predictions > 0.5).astype(int).flatten()

# Get true labels
y_true = test_gen.classes

# Generate Confusion Matrix
conf_matrix = confusion_matrix(y_true, y_pred_classes)
class_names = list(train_gen.class_indices.keys())

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Generate Classification Report
print("\n[STEP 7.3] Classification Report:")
print(classification_report(y_true, y_pred_classes, target_names=class_names))
```

```python
## 📋 Prediction Analysis Table

print("\n[STEP 8] Generating Prediction Analysis Table...")

# --- Prediction Details ---
# 1. Get filenames
filenames = test_gen.filenames
# 2. Get true labels (y_true already calculated in Step 7)
# 3. Get predicted classes (y_pred_classes already calculated in Step 7)
# 4. Get confidence scores (predictions already calculated in Step 7)
confidence_scores = predictions.flatten()

results_df = pd.DataFrame({
    'Filename': filenames,
    'True_Label_Index': y_true,
    'Predicted_Label_Index': y_pred_classes,
    'Confidence_Score': confidence_scores
})

# Map indices back to class names
class_map = {v: k for k, v in train_gen.class_indices.items()}
results_df['True_Label'] = results_df['True_Label_Index'].map(class_map)
results_df['Predicted_Label'] = results_df['Predicted_Label_Index'].map(class_map)
results_df['Is_Correct'] = results_df['True_Label'] == results_df['Predicted_Label']

# Calculate final confidence percentage (confidence in the predicted class)
def final_confidence(row):
    pred = row['Confidence_Score']
    # If predicted as FAKE (1), confidence is pred * 100
    if row['Predicted_Label_Index'] == 1:
        return pred * 100
    # If predicted as REAL (0), confidence is (1 - pred) * 100
    else:
        return (1 - pred) * 100

results_df['Final_Confidence_%'] = results_df.apply(final_confidence, axis=1).round(2)


# --- Display Tables ---

# 1. Incorrect Predictions (Misclassified)
misclassified_df = results_df[results_df['Is_Correct'] == False].sort_values(by='Final_Confidence_%', ascending=False)
print("\n--- Top 10 Misclassified Images (High Confidence Errors) ---")
display(misclassified_df[['Filename', 'True_Label', 'Predicted_Label', 'Final_Confidence_%']].head(10))

# 2. Correct Predictions (Easiest to classify)
correct_df = results_df[results_df['Is_Correct'] == True].sort_values(by='Final_Confidence_%', ascending=False)
print("\n--- Top 10 Correctly Classified Images (Highest Confidence) ---")
display(correct_df[['Filename', 'True_Label', 'Predicted_Label', 'Final_Confidence_%']].head(10))

# 3. Overall Summary
print("\n--- Overall Test Set Prediction Summary ---")
summary_df = results_df.groupby(['True_Label', 'Predicted_Label']).agg(Count=('Filename', 'size')).reset_index()
summary_df['Percentage'] = summary_df['Count'] / results_df['Filename'].count() * 100
display(summary_df)

print("\n✅ Analysis Complete!")
```

```python
#Step 9: Prediction on New Images
import tensorflow as tf
import cv2
import numpy as np
import os

# --- Configuration from previous steps ---
MODEL_NAME = 'best_simple_cnn_deepfake_model.h5'
IMG_SIZE = (224, 224)

# --- Prediction Function ---
def predict_new_image(image_path, model_name, img_size=IMG_SIZE, threshold=0.5):
    """
    Loads a trained model and predicts if the given image is Real or Fake.
    """
    if not os.path.exists(image_path):
        return f"Error: Image file not found at '{image_path}'", 0.0

    # 1. Load the trained model
    try:
        model = tf.keras.models.load_model(model_name)
    except Exception as e:
        print(f"Error loading model: {e}")
        return "Error: Model loading failed", 0.0

    # 2. Load and Preprocess Image
    img = cv2.imread(image_path)
    if img is None:
        return "Error: Could not read image file", 0.0

    # Resize and convert to RGB (from BGR)
    img = cv2.resize(img, img_size)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Normalize pixel values (Scale to 0-1, as done by the generators)
    img = img / 255.0

    # Add batch dimension (model expects a batch of images, even if it's just one)
    img = np.expand_dims(img, axis=0)

    # 3. Make Prediction
    # The output is a single probability score (0 to 1) for the 'Fake' class
    probability = model.predict(img, verbose=0)[0][0]
    
    # 4. Determine Label and Confidence
    if probability >= threshold:
        label = 'FAKE'
        # Confidence is how sure the model is that it's FAKE
        confidence = probability * 100
    else:
        label = 'REAL'
        # Confidence is how sure the model is that it's REAL (1 - probability)
        confidence = (1 - probability) * 100

    # 5. Visualize and Print Result
    plt.figure(figsize=(4, 4))
    plt.imshow(img[0]) # Display the pre-processed image
    plt.title(f"Prediction: {label} ({confidence:.2f}%)", fontsize=12, fontweight='bold')
    plt.axis('off')
    plt.show()

    return label, confidence

# --- Example Usage ---
print("\n" + "="*60)
print("PREDICTION EXAMPLE: Real/Fake Classification")
print("="*60)

# ⚠️ IMPORTANT: Change this path to a real image file on your computer
TEST_IMAGE_PATH = r"C:\Users\adarsh\OneDrive\Documents\GitHub\DEEP_FAKE_DETECTION\data_set\Validation\Real\real_97.jpg"

# Ensure you have a placeholder image to demonstrate the output structure if needed
# If you don't have a specific test image yet, use a placeholder path, but the prediction will fail.

# Example: Run the prediction
predicted_label, conf = predict_new_image(TEST_IMAGE_PATH, MODEL_NAME)

print(f"\nImage Path: {TEST_IMAGE_PATH}")
print(f"--- Model Prediction ---")
print(f"Label: {predicted_label}")
print(f"Confidence: {conf:.2f}%")
print("="*60)
```
